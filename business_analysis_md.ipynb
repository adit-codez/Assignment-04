B1. Problem Formulation

(a) Machine learning problem
This is a supervised regression problem because we want to predict how many items will be sold, which is a number.
The target variable is items sold.
The input features can be store size, store location, promotion type, competition density, customer demographics, weekend/festival flags, and monthly timing.
Regression is the right choice because the output is continuous, not just yes/no or one class out of many.

(b) Why items sold is better than revenue
Using items sold is better because revenue can change because of price changes, discounts, or expensive products, but items sold directly shows how well the promotion worked.
If a promotion makes people buy more units, that is a clearer sign of success.
This shows an important ML idea: the target variable should match the real business goal.

(c) Better modelling strategy than one global model
Instead of using one single model for all 50 stores, I would use a segmented approach.
For example, I would group stores by location type like urban, semi-urban, and rural, and maybe also by store size.
This is better because different stores do not behave the same way, so separate or segment-based models can learn local patterns more accurately.


B2. Data and EDA Strategy

(a) How to join the tables
I would join all four tables using the common keys like store ID, promotion ID, and date.
The final dataset should have one row per store per month per promotion.
Before modelling, I would aggregate transaction data into monthly store-level values, such as total items sold, average competition density if needed, and calendar flags like weekend or festival.
This gives one clean record for each store-month-promotion combination.

(b) EDA I would do
First, I would plot items sold by promotion type to see which promotion works best.
Second, I would compare sales across urban, semi-urban, and rural stores to see location differences.
Third, I would check monthly trends to see if sales go up during festivals or certain seasons.
Fourth, I would look at competition density vs items sold to see if more competition lowers sales.
These charts would help me decide what features to make and whether I need separate models for different store groups.

(c) Effect of 80% no-promotion cases
If 80% of the data has no promotion, the model may learn that “no promotion” is the most common pattern and pay less attention to the real promotions.
That can make the model biased and less useful.
To handle this, I would check the distribution carefully, use class balancing ideas if needed, and make sure evaluation is done properly across all promotion types.

B3. Model Evaluation and Deployment

(a) Train-test split and metrics
Because the data is monthly and time-based, I would split it in order.
I would train on the first 80% of months and test on the last 20% of months.
A random split is bad here because it can mix future data into training, which is not realistic.
For evaluation, I would use:
RMSE, to see how far predictions are from the real values.
MAE, because it is easy to understand in terms of average error.
R-squared, to see how much of the variation in sales the model explains.

(b) Why the model recommends different promotions for the same store
If the model recommends Loyalty Points in December but Flat Discount in March for the same store, it is probably because the month is different and the customer behavior is different too.
I would check feature importance to see which features matter most, like month, festival flag, competition, and location type.
Then I would explain to the marketing team that the model is not just looking at the store, but also the season and situation around that store.

(c) Deployment process
First, I would save the trained model using something like joblib or pickle.
Then, every month, I would prepare the new store data using the same steps as training, like encoding and scaling, and feed it into the saved model.
The model would then give promotion recommendations for all 50 stores.
For monitoring, I would watch if prediction errors start increasing or if the new data looks very different from the training data.
If performance drops or data changes a lot, I would retrain the model using fresh data.
This way, the model stays useful over time.
